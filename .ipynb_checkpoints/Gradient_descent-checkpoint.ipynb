Functions

```python
def objective_function(G, T, C1, C2):
    return C1 * G + C2 * G * T
```

```python
def loss_function_dc(G, T, C1, C2, C3, C4, C5):
    return C1 * T * T + C2 * G * T + C3 * G + C4 * G * G + C5
```

```python
def gradient_descent_ir(ir, temp, y_target, C1, C2, learning_rate, num_iterations):
    for i in range(num_iterations):
        # Calculate the current value of c
        y_current = objective_function(ir, temp, C1, C2)
        
        # Calculate the gradients
        gradient = C1 + C2 * temp
   
        # Update a and b using the gradients
        ir -= learning_rate * gradient * (y_current - y_target)
        
        # Check if the calculated c is close enough to the target c
        if abs(y_current - y_target) < 1e-5:
            break
    
    return ir
```

```python
def gradient_descent_t(ir, temp, y_target, C1, C2, learning_rate, num_iterations, temp_max, temp_min):
    for i in range(num_iterations):
        # Calculate the current value of c
        y_current = objective_function(ir, temp, C1, C2)
        
        # Calculate the gradients
        gradient = C2 * ir
        
        # Update a and b using the gradients
        temp -= learning_rate * gradient * (y_current - y_target)
        
        temp = max(temp_min, min(temp, temp_max))
        
        # Check if the calculated c is close enough to the target c
        if abs(y_current - y_target) < 1e-5:
            break
    
    return temp
```

```python
mask_sp_max_ghi = sp_env_df['ir'] > max_ghi
mask_sp_min_ghi = sp_env_df['ir'] < min_ghi
mask_sp_max_temp = sp_env_df['ambient_temp2'] > max_temp
mask_sp_min_temp = sp_env_df['ambient_temp2'] < min_temp
mask_sp_max_rh = sp_env_df['ambient_rh'] > max_rh
mask_sp_min_rh = sp_env_df['ambient_rh'] < min_rh
mask_ae_max_ghi = ae_df['GHI'] > max_ghi
mask_ae_min_ghi = ae_df['GHI'] < min_ghi

sp_env_df.loc[mask_sp_max_ghi, 'ir'] = np.nan
sp_env_df.loc[mask_sp_min_ghi, 'ir'] = np.nan
sp_env_df.loc[mask_sp_max_temp, 'ambient_temp2'] = np.nan
sp_env_df.loc[mask_sp_min_temp, 'ambient_temp2'] = np.nan
sp_env_df.loc[mask_sp_max_rh, 'ambient_rh'] = np.nan
sp_env_df.loc[mask_sp_min_rh, 'ambient_rh'] = np.nan
ae_df.loc[mask_ae_max_ghi, 'GHI'] = np.nan
ae_df.loc[mask_ae_min_ghi, 'GHI'] = np.nan

sp_env_df['ir'] = sp_env_df['ir'].fillna(method='ffill')
sp_env_df['ambient_temp2'] = sp_env_df['ambient_temp2'].fillna(method='ffill')
sp_env_df['ambient_rh'] = sp_env_df['ambient_rh'].fillna(method='ffill')
ae_df['GHI'] = ae_df['GHI'].fillna(method='ffill')
```

```python
ws_df['weather_points'] = 0
for i in range(ws_df.shape[0]):
    points = 0
    for k in str(ws_df['weather_condition'][i]).split("/"):
        for kk in k:
            if(kk in severe_weather):
                points = points + 10
            elif(kk in mild_weather):
                points = points + 5
    ws_df.loc[i, 'weather_points']= points
```

Linear Regression to calculate P_AC = C1 * G + C2 * G * T + C3

```python
import numpy as np
from sklearn.linear_model import Ridge

P_AC = []
G = []
T = []

for i in range(total_days):
    current_time = start_date + timedelta(days=i)
    P_AC.append(sp_op_df.iloc[sp_start_index[i]:sp_start_index[i]+sp_num_record[i]]['ac_power'])
    G.append(ae_df.iloc[ae_start_index[i]:ae_end_index[i]:interval]['GHI'])
    T.append(sp_env_df.iloc[sp_start_index[i]:sp_start_index[i]+sp_num_record[i]]['ambient_temp2'])
    
P_AC = pd.concat(P_AC)  
T = np.array(pd.concat(T))
G = np.array(ae_df.loc[0:ae_df.shape[0]:interval, 'GHI'])

print(P_AC.shape, T.shape, G.shape)

# Constructing the design matrix
X = np.column_stack((G, G * T, np.ones_like(G)))

ridge_model = Ridge(alpha=1.0)  # Adjust alpha for regularization strength
ridge_model.fit(X, P_AC)  # X is the design matrix from previous example

# Get coefficients
AC_C1 = ridge_model.coef_[0]
AC_C2 = ridge_model.coef_[1]
AC_C3 = ridge_model.coef_[2]

#print("Coefficient C_1:", C_1)
#print("Coefficient C_2:", C_2)
#print("Coefficient C_3:", C_3)

AC_C1, AC_C2, AC_C3
```

#### Calculate Similarities

```python
# GD Parameters
learning_rate = 0.001
num_iterations = 10000 

# Display Parameters
num_display = 10

similarity = np.zeros((total_days, 5))

for i in range(total_days):
    current_time = start_date + timedelta(days=i)
    
    ir = ae_df.iloc[ae_start_index[i]:ae_end_index[i]:interval]['GHI'].values
    temp = sp_env_df.iloc[sp_start_index[i]:sp_start_index[i]+sp_num_record[i]]['ambient_temp2'].values
    power = sp_op_df.iloc[sp_start_index[i]:sp_start_index[i]+sp_num_record[i]]['ac_power'].values
    
    time_series_minute = sp_env_df.iloc[sp_start_index[i]:sp_start_index[i]+sp_num_record[i]]['time'].values


    # Before                                                     
    value = objective_function(ir, temp, AC_C1, AC_C2)
  
    similarity[i, 0] = evaluate_similarity(value, power)
    #print("Power before IR:",evaluate_similarity(value, power))

    updated_ir =  np.zeros_like(ir)
    for idx in range(ir.shape[0]):
        updated_ir[idx] = gradient_descent_ir(ir[idx], temp[idx], power[idx], AC_C1, AC_C2, learning_rate, num_iterations)

    updated_value_ir = objective_function(updated_ir, temp, AC_C1, AC_C2)
    similarity[i, 1] = evaluate_similarity(updated_value_ir, power)
    #print("Power after IR:",evaluate_similarity(updated_value_ir, power))
    similarity[i, 2] = evaluate_similarity(ir, updated_ir)
    #print("Ir:",evaluate_similarity(ir, updated_ir))

    # Temp Gradient Descent

    updated_temp = np.zeros_like(temp)
    for idx in range(temp.shape[0]):
        updated_temp[idx] = gradient_descent_t(updated_ir[idx], temp[idx], power[idx], AC_C1, AC_C2, learning_rate, num_iterations, max_temp, min_temp)

    updated_value_t = objective_function(updated_ir, updated_temp, AC_C1, AC_C2)
    similarity[i, 3] = evaluate_similarity(updated_value_t, power)
    #print("Power after T:",evaluate_similarity(updated_value_t, power))
    similarity[i, 4] = evaluate_similarity(temp, updated_temp)
    #print("Temp:", evaluate_similarity(temp, updated_temp))
```
